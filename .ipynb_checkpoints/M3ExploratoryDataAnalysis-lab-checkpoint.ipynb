## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
df['ConvertedComp'].plot(kind='box', vert=False)
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
import numpy as np
abc=df['ConvertedComp'].dropna()
q1=np.percentile(abc,25)
q3=np.percentile(abc,75)
iqr= q3 - q1
iqr
```

Find out the upper and lower bounds.


```python
upper_bound = q3 + 1.5 * iqr
lower_bound = max(0, q1 - 1.5 * iqr)  # Set lower bound to zero if it's less than zero

print(f"Upper Bound: {upper_bound:.2f}")
print(f"Lower Bound: {lower_bound:.2f}")
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
outliers = df[(df['ConvertedComp'] > upper_bound) | (df['ConvertedComp'] < lower_bound)]

num_outliers = len(outliers)
print(f"Number of outliers in ConvertedComp: {num_outliers}")
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
df_no_outliers = df[(df['ConvertedComp'] >= lower_bound) & (df['ConvertedComp'] <= upper_bound)]

df_no_outliers['ConvertedComp'].plot(kind='box', vert=False)
```
